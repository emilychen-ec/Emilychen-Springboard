### Datetime

Add column `publish_day` to show which days of the week the videos were published

```python

US_data2['publish_day']=US_data2['publish_time'].apply(lambda x: datetime.strptime(x[:10], "%Y-%m-%d").date().strftime('%a'))

US_data2.head()
```

Addition of column "Days before Trend".
First of all, I transform "trending_date" and "publish_time" column to datetime format.Then I calculate the date differences using the subtraction operation.

```python
publish_time = pd.to_datetime(US_data2['publish_time'], format ='%Y-%m-%dT%H:%M:%S.%fZ')

US_data2['publish_date']=publish_time.dt.date
US_data2['publish_hour']=publish_time.dt.hour
US_data2['publish_year']=publish_time.dt.year

US_data2['publish_hour'] = US_data2['publish_hour'].astype(object)
US_data2['publish_year'] = US_data2['publish_year'].astype(object)
```

```python
US_data_drop = US_data2.copy()

US_data_first = US_data_drop.drop_duplicates(subset=['video_id'], keep = 'first', inplace = False)

US_data_first.head()
```

```python
US_data_first['trending_date'] = pd.to_datetime(US_data_first['trending_date'], format='%y.%d.%m').dt.date
#US_data_first['Days_before_trend'] = (US_data_first.trending_date - US_data_first.publish_date)/ np.timedelta64(1, 'D')

US_data_first['Days_before_trend'] = (US_data_first.apply(lambda x: (x['trending_date']) - (x['publish_date']), axis=1)) /np.timedelta64(1, 'D')
US_data_first['Days_before_trend'] = US_data_first['Days_before_trend'].astype(int)

US_data_first.tail(3)
```

```python
US_data_first.shape
```

Addition of column 'Views/day before trending'

```python
US_data_first['Days_before_trend'].min()
```

```python
 US_data_first[US_data_first['Days_before_trend'] ==0].head(3)
```

```python
#US_data_first_copy['new_col'] = US_data_first.apply(lambda x: x.views/x.Days_before_trend if x.Days_before_trend != 0 else x.views, axis = 1)

US_data_first.loc[:,'Views/days before trending'] = US_data_first.apply(lambda x: x.views/x.Days_before_trend if x.Days_before_trend != 0 else '', axis = 1)
US_data_first.loc[:,'likes/days before trending'] = US_data_first.apply(lambda x: x.likes/x.Days_before_trend if x.Days_before_trend != 0 else '', axis = 1)
US_data_first.loc[:,'dislikes/day before trending'] = US_data_first.apply(lambda x: x.dislikes/x.Days_before_trend if x.Days_before_trend != 0 else '', axis = 1)
US_data_first.loc[:,'comment_count/day before trending'] = US_data_first.apply(lambda x: x.comment_count/x.Days_before_trend if x.Days_before_trend != 0 else '', axis = 1)
```

```python
US_data_first[US_data_first['Days_before_trend'] == 0].head(3)
```

### Visualization for Bi-variate

```python
def scatterplots(columns, ncol=None, figsize=(20, 15)):
    if ncol is None:
        ncol = len(columns)
    nrow = int(np.ceil(len(columns) / ncol))
    fig, axes = plt.subplots(nrow, ncol, figsize = figsize, squeeze=False)
    fig.subplots_adjust(wspace=0.5, hspace=1)
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        ax.scatter(x = col, y = 'likes', data=US_data_final, alpha=0.5)
        ax.set(xlabel=col, ylabel='likes')
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```

```python
features = [i for i in US_data_final_num.columns if i not in ['likes']]
```

```python
scatterplots(features, ncol=4, figsize=(15, 15))
plt.rcParams.update({'font.size': 12})
```
