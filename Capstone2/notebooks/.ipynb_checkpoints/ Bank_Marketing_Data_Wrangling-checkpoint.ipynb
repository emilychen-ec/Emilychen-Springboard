### Number Of Missing Values By Column
Count the number of missing values in each column and sort them.

```python
missing = pd.concat([bank.isnull().sum(), 100*bank.isnull().mean()], axis=1)
missing.columns = ['count', '%']
missing.sort_values(by='count')
```

### Outliers

```python
def Box_plots(df):
    plt.figure(figsize=(10, 4))
    plt.title("Box Plot")
    sns.boxplot(df)
    plt.show()
    

Box_plots(bank['age'])
```

```python
fig, ax = plt.subplots(3, 2, figsize=(20,15))
bank.age.plot(kind='box', ax=ax[0,0])
ax[0,0].set_title('age', fontweight="bold", size=20)

bank.balance.plot(kind='box', ax=ax[0,1])
ax[0,1].set_title('balance', fontweight="bold", size=20)

bank.duration.plot(kind='box', ax=ax[1,0])
ax[1,0].set_title('duration', fontweight="bold", size=20)

bank.day.plot(kind='box', ax=ax[1,1])
ax[1,1].set_title('day', fontweight="bold", size=20)

bank.campaign.plot(kind='box', ax=ax[2,0])
ax[2,0].set_title('campaign ', fontweight="bold", size=20)

bank.pdays.plot(kind='box', ax=ax[2,1])
ax[2,1].set_title('pdays', fontweight="bold", size=20)

plt.subplots_adjust(wspace=0.5,hspace=1.2);
```
