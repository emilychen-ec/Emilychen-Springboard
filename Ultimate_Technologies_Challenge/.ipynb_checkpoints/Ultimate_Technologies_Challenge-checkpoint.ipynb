### Part 1 - Exploratory data analysis 
***

```python
login_info = pd.read_json('logins.json')
login_info
```

```python
#Find out the time span of data
login_info['login_time'].min(), login_info['login_time'].max()
```

```python
#Create a new dataframe with timestamp counts
login_counts = login_info.login_time.value_counts().rename_axis().to_frame('counts')
login_counts.sort_index()
```

```python
#Aggregate these login counts based on 15minute time intervals
login_delta15=login_counts.resample('15min').sum()
login_delta15
```

```python
#visualize the resulting time series of login counts
rcParams['figure.figsize'] = 18, 8
decomposition = sm.tsa.seasonal_decompose(login_delta15, period = 4*24, model='addictive')
fig = decomposition.trend.plot()
plt.show()
```

login count a slight upward trend from Jan to April. 

```python
week_sum_df=login_counts.groupby(login_counts.index.dayofweek).sum().reset_index().rename(columns={"index": "day", "counts": "login_sum"})
week_sum_df['day']=week_sum_df['day'].replace({0:'Mon',1:'Tues', 2:'Wed', 3:'Thurs', 4:'Fri', 5:'Sat', 6:'Sun'})
week_sum_df['percent'] = week_sum_df['login_sum']/week_sum_df['login_sum'].sum()

week_sum_df.plot(x='day', y='percent', kind='bar', 
        title='Login Percent by days of week')
plt.xticks(rotation=45)
plt.show()
```

Weekends, espcially Saturdays has the most login counts. There is a upward trend from Monday to Friday. 

```python
hour_sum_df=login_counts.groupby(login_counts.index.hour).sum().reset_index().rename(columns={"index": "hour", "counts": "login_sum"})
hour_sum_df['percent'] = hour_sum_df['login_sum']/hour_sum_df['login_sum'].sum()

hour_sum_df.plot(x='hour', y='percent', kind='bar', 
        title='Login Percent by Hour')
plt.xticks(rotation=45)
plt.show()
```

We see the most login counts from 21:00 -4:00. 11:00 and 12:00 also has a high login counts. The lowest login counts is from 6:00-9:00 in the mornings.

#### Machine Learning Model

```python
#`avg_surge` and `surge_pct` is highly correlated from the heatmap. I will delete `avg_surge` for Modeling
X_modeling = df_n_corr.drop(['avg_surge'],axis=1)
X_modeling.info()
```

```python
#Convert categorical variables into dummy variables
categorical_features = ['city', 'phone', 'signup_day','ultimate_black_user']
df_n_corr_dummies = pd.get_dummies(X_modeling, columns = categorical_features)
print('The data have ', X_modeling.shape[0], ' rows and ', X_modeling.shape[1], ' columns\n')
print('column names: \n')
print('\n'.join(list(df_n_corr_dummies.columns)))
```

```python
X= df_n_corr_dummies.drop(columns = ['retained'])
y = df_n_corr_dummies['retained']
Xtrain,Xtest, ytrain, ytest = train_test_split(X,y, test_size=.3, random_state=8)
```

#### XGB Model


```python
XGBModel =XGBClassifier(objective = 'binary:logistic',random_state = 2)
parameters = {
    "n_estimators":[10,20],
    "learning_rate": [0.1, 0.25, 0.5,0.75,1],
    "max_depth":[2,5,8],
    'gamma':[0,1, 2] 
    }
#Hyperparameter search using GridSearchCV
XGBModel= GridSearchCV(XGBModel, parameters, cv= 5,scoring='roc_auc')
XGBModel.fit(Xtrain, ytrain)
print(XGBModel.best_params_, XGBModel.best_score_)
```

```python
XGBModel.score(Xtrain, ytrain)
```

There is 16% missing data in ‘avg_rating_of driver’ and small percent of missing data in ‘avg_rating_by_driver’ and phone’. I decided to choose XGB model which can handle np.nan without imputation. 

Alternatively, I considered GB model, but it requires me to impute the nan data. Random forest model is nice but it requires too many computational capacity. 

Roc_Auc scoring is a better indicator of model performance because we are dealing with the imbalanced data and Roc_Auc measures the model sensitivity and specificity.

```python

xgb_fpr, xgb_tpr, thresholds = roc_curve(ytrain, XGBModel.best_estimator_.predict_proba(Xtrain)[:, 1])

df_fpr_tpr = pd.DataFrame({'FPR':xgb_fpr, 'TPR':xgb_tpr, 'Threshold':thresholds})

df_fpr_tpr.head()
```

```python
# Calculate the G-mean for the optimal threshold
gmean = np.sqrt(xgb_tpr * (1 - xgb_fpr))

# Find the optimal threshold
index = np.argmax(gmean)
thresholdOpt = round(thresholds[index], ndigits = 4)
gmeanOpt = round(gmean[index], ndigits = 4)
fprOpt = round(xgb_fpr[index], ndigits = 4)
tprOpt = round(xgb_tpr[index], ndigits = 4)
print('Best Threshold: {} with G-Mean: {}'.format(thresholdOpt, gmeanOpt))
print('xgb_fpr: {}, xgb_tpr: {}'.format(fprOpt, tprOpt))
```

The optimal threshold is 0.358. It has an AUC value of 0.85 in distinguishing between the positive and negative classes. Since the dataset is highly imbalanced, I used the predict_probability >= threshold in the confusion matrix to handle the imbalanced data.


```python
# Plot ROC curve
plt.figure(figsize=(12,8))
plt.plot(xgb_fpr, xgb_tpr,lw=2)
#Plot the default threshold
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
#Plot the optimal threshold
plt.plot([0.22],[0.78],marker ='o', markersize=10, label='Optimal threshold 0.358')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.legend()
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-AUC comparison')
plt.show()
```
