```python
US_data.describe(include = 'object')
```

Here is an example of one video appeared more than once on the trending list but with different titles and descriptions.

```python
grouped = US_data.groupby("video_id")
groups = []
wanted_groups = []
for key, item in grouped:
    groups.append(grouped.get_group(key))

for g in groups:
    if len(g['title'].unique()) != 1:
        wanted_groups.append(g)

wanted_groups[0]
```

Addition of column "Days before Trend".
First of all, I transform "trending_date" and "publish_time" column to datetime format.Then I calculate the date differences using the subtraction operation.

```python
US_data['trending_date'] = pd.to_datetime(US_data['trending_date'], format='%y.%d.%m').dt.date

publish_time = pd.to_datetime(US_data['publish_time'], format ='%Y-%m-%dT%H:%M:%S.%fZ')
US_data['publish_date']=publish_time.dt.date
US_data['publish_time']=publish_time.dt.time
US_data['publish_hour']=publish_time.dt.hour
```

```python
US_data.head(2)
```

```python
US_data['Days_before_trend'] = (US_data.trending_date - US_data.publish_date)/ np.timedelta64(1, 'D')
US_data['Days_before_trend'] = US_data['Days_before_trend'].astype(int)
US_data.tail(3)
```

Addition of the category column from US_category_id.json file

```python
categories = {category['id']:category['snippet']['title'] for category in US_data_categories['items']}
US_data.insert(5, 'category', US_data['category_id'].astype(str).map(categories))
```
